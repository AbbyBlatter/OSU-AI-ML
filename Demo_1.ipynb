##Graphing:

The first graph displays each marvel movie and how much it made worldwide.

```python
try:
  df.plot.bar(x='movie', y='worldwide gross ($m)')

except:
  print("Could not create graph.")

#NOTE: "scatter" in df.plot can be change to any chart type, for example, pie to produce pie graph, bar, histogram, line, etc.
```

The second graph displays the mean worldwide gross earnings across categories by averaging the gross made from each category of movies




```python
category_success = df.groupby('category')['worldwide gross ($m)'].mean().sort_values(ascending=False)
plt.figure(figsize=(9, 9))
sns.barplot(x=category_success.index, y=category_success.values)
plt.xticks(rotation=45, ha='right')
plt.xlabel('Category')
plt.ylabel('Mean Worldwide Gross ($m)')
plt.title('Category vs Mean WW Gross')
try:
  plt.show()
except:
  print("Could not create graph.")
```

The third graph displays the budget across categories by averaging the budget from each category of movies.


```python
category_success = df.groupby('category')['budget'].mean().sort_values(ascending=False)
plt.figure(figsize=(9, 9))
sns.barplot(x=category_success.index, y=category_success.values)
plt.xticks(rotation=45, ha='right')
plt.xlabel('Category')
plt.ylabel('Budget')
plt.title('Category vs Budget')
try:
  plt.show()
except:
  print("Could not create graph.")
```

The last 2 graphs helps describe the correlation with the category of movie made and how much it makes. As shown Avengers movies generate the most revenue for Marvel. The next steps would be to then analyze the profits made from each category to see if the revenue was worth the budget.
